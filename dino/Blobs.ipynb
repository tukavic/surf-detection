```python
path = 'video_frames'
for f in os.listdir(path):
    img = cv2.imread(path + '/' + f)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    blobs = blob_doh(cv2.cvtColor(img, cv2.COLOR_RGB2GRAY), threshold=0.008)
    plt.imshow(img)
    plt.scatter(blobs[:, 1], blobs[:, 0], c='white', marker='^', s=75)
    plt.show()
```
